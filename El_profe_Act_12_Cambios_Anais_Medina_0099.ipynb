## Las variables
1.  CP = --> Costo de producto
2.  CD = --> Cantidad de dinero
3.  Cambio

## Formulas
$Cambio = CD - CP$

## Diagrama de flujo lucidchart
<img src="https://raw.githubusercontent.com/AnaisMedinaZubia/Calcular-cambio-if-else/refs/heads/main/DF%20if%20else%20cambio%20ye.png" width="400" height="500" alt="Texto descriptivo">



```python
# Para cuando sea verdadero(Si)
# calculando el cambio usando if else
# Datos de entrada CD, CP
# Mostrar cambio si es falso
# si es verdad mostrar No alcanza
CP = int(input("Ingrese la costo de producto: "))
CD = int(input("Ingrese el cantidad del dinero: "))

if CP > CD:
    print("No alcanza el dinero")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)

print("Fin del programa")
print("Programa realizado por Anais Medina NL = 36")
```

```python
# Para cuando sea falso(No)
# calculando el cambio usando if else
# Datos de entrada CD, CP
# Mostrar cambio si es falso
# si es verdad mostrar No alcanza
CP = int(input("Ingrese la costo de producto: "))
CD = int(input("Ingrese el cantidad del dinero: "))

if CP > CD:
    print("No alcanza el dinero")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)

print("Fin del programa")
print("Programa realizado por Anais Medina NL = 36")
```
